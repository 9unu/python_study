택시기사에게 50명의 승객과 매칭 기회
총 탑승 승색 수는?

조건 1. 승객 별 운행 소요 시간 : 5~50분 (난수)
조건 2. 소요시간 5분~15분 사이의 승객만 매칭

```python
from random import randint
dic=dict()
for i in range(51):
    dic[i]=randint(5,51)
count =0
for i in range(51):
    if(dic[i]>=5 and dic[i]<=15):
        print(f"[o]{i}번째 손님 (소요시간 : {dic[i]}분)")
        count+=1
    else:
        print(f"[x]{i}번째 손님 (소요시간 : {dic[i]}분)")

print(count)
```
